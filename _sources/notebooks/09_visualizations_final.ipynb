## Limitations & Data Granularity

This analysis is **system-level** in scope. CAISO metrics (net load, curtailment, wholesale prices) are aggregate California-wide values — they do not vary by ZIP code or utility territory. This is a fundamental constraint of publicly available grid operations data.

**Why ZIP-level adoption doesn't predict ZIP-level grid stress:**

1. **Aggregate effect dominates:** When solar ramps down at sunset across all of California, CAISO sees a single system-level event. Whether solar is concentrated in Riverside or spread across 100 ZIPs, the system ramp is the same.

2. **Utility-scale solar drives the curve:** The duck-curve deepening is primarily driven by utility-scale solar (Mojave Desert, Central Valley projects), not distributed BTM solar.

3. **Data granularity mismatch:** System-level CAISO metrics are invariant across all ZIPs within the CAISO interconnection — the same ramp magnitude is assigned to every ZIP.

**To test local effects, we would need:**
- Feeder-level voltage and frequency data (not publicly available)
- Distribution-network operator data (requires utility data-sharing agreements)
- Customer-level interval metering data (requires Green Button access)
